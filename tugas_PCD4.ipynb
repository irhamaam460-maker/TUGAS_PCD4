```python
# ============================
# IMPORT LIBRARY
# ============================
import cv2
import numpy as np
import matplotlib.pyplot as plt

# ============================
# UPLOAD IMAGE
# ============================
from google.colab import files
uploaded = files.upload()

# Ambil nama file
image_path = list(uploaded.keys())[0]

# Baca gambar
img = cv2.imread(image_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Tampilkan gambar asli
plt.imshow(img_rgb)
plt.title("Gambar Asli")
plt.axis('off')
plt.show()


# ============================
# 1. RGB -> GRAYSCALE
# ============================
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
plt.title("Grayscale")
plt.axis('off')
plt.show()


# ============================
# 2. GRAYSCALE -> BINER
# ============================
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

plt.imshow(binary, cmap='gray')
plt.title("Biner")
plt.axis('off')
plt.show()


# ============================
# 3. GRAYSCALE -> MBIT (Quantization)
# ============================
# contoh 4-bit (16 level)
mbit = np.floor(gray / 16) * 16

plt.imshow(mbit, cmap='gray')
plt.title("Grayscale ke 4-bit")
plt.axis('off')
plt.show()


# ============================
# 4. BRIGHTNESS
# ============================
brightness = cv2.convertScaleAbs(gray, alpha=1, beta=50)

plt.imshow(brightness, cmap='gray')
plt.title("Brightness +50")
plt.axis('off')
plt.show()


# ============================
# 5. CONTRAST
# ============================
contrast = cv2.convertScaleAbs(gray, alpha=1.5, beta=0)

plt.imshow(contrast, cmap='gray')
plt.title("Contrast x1.5")
plt.axis('off')
plt.show()


# ============================
# 6. HISTOGRAM GRAYSCALE
# ============================
plt.hist(gray.ravel(), bins=256, range=[0,256])
plt.title("Histogram Grayscale")
plt.show()


# ============================
# 7. HISTOGRAM EQUALIZATION
# ============================
equalized = cv2.equalizeHist(gray)

plt.imshow(equalized, cmap='gray')
plt.title("Histogram Equalization")
plt.axis('off')
plt.show()

# Histogram setelah equalization
plt.hist(equalized.ravel(), bins=256, range=[0,256])
plt.title("Histogram Setelah Equalization")
plt.show()
```
